```python
# use online dataset: iris

# https://archive.ics.uci.edu/ml/datasets/iris

# pandas import
import pandas as pd

# read csv file
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data',
                    header=None,
                    names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class'])

# print first 5 rows

display(df.head())
# print last 5 rows

display(df.tail())
```

```python
print(list(df['sepal_length'] > 5))

sepal_length_bigger_than_5 = filter(lambda x: x > 2, df['sepal_length'])

display(list(sepal_length_bigger_than_5))
```

```python
# map sepal_length to sepal_length + 1
sepal_length_plus_1 = map(lambda i: i + 1, df['sepal_length'])
print(list(sepal_length_plus_1))
```

```python
def filter_func(x):
    return x > 5

sepal_length_bigger_than_5 = filter(filter_func, df['sepal_length'])

display(list(sepal_length_bigger_than_5))
```

```python
# currying

def filter_func(x, y, z):
    return x > y + z

# currying
def filter_func(x):
    def inner(y, z):
        return x > y + z
    return inner

filter_func_5 = filter_func(5)
sepal_length_bigger_than_5 = filter(lambda x: filter_func_5(x, 2), df['sepal_length'])

display(list(sepal_length_bigger_than_5))
# filter sepal_length > 5
sepal_length_bigger_than_5 = filter(lambda x: x > 5, df['sepal_length'])
display(list(sepal_length_bigger_than_5))
```
